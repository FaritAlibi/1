```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#функция loss

def func(x):
    return x**3 - 6*x**2 - 15*x + 10
```

```python
#функция производной

def func_derivative(x):
    return 3*x**2 - 12 * x - 15
```

```python
#функция град.спуска с рисованием графиков

def descent(x_start, learning_rate, threshold):
    x_start = x_start
    learning_rate = learning_rate
    threshold = threshold
    
    x_next = x_start - learning_rate*func_derivative(x_start)

    fig = plt.figure()
    fig.set_size_inches(8,6)

    plt.plot(np.linspace(-5,10,150), func(np.linspace(-5,10,150)), alpha=0.5)
    
    while abs(x_start - x_next):
        x_start = x_next
        x_next = x_start - learning_rate*func_derivative(x_start)

        plt.scatter([x_start, x_next], 
                    [func(x_start), func(x_next)],
                    color='green',
                    s=80)
        plt.plot([x_start, x_next], 
                    [func(x_start), func(x_next)],
                    color='red')
            
    print(x_next)
```

```python
descent(10, 0.01, 0.0001)
```
